```python
# Import required packages
import numpy as np
import pandas as pd
```

Carregando o conjunto de dados em um wuadro de dados usando Pandas

```python
#@title
df=pd.read_csv('agricultural_raw_material.csv')
```

Explorando o conjunto de dados

```python
df.info
#checking Null Values of each column
df.isnull().sum()
```

Lidar com tratos ausentes,incorretos e inv√°lidos

```python
# Replacing %, "," and "-"
df = df.replace('%', '', regex = True)
df = df.replace(',', '', regex = True)
df = df.replace('-', '', regex = True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
# Dropping rows with NaN values
df = df.dropna()
# Check to see if all NaN values are resolved
df.isnull().sum()
# Converting data type to float
lst = ["Coarse wool Price", "Coarse wool price % Change","Copra Price", "Copra price % Change", "Cotton price % Change", "Fine wool Price", "Fine wool price % Change", "Hard log price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
df.head()
```

Coluna de datas e horacolunas:
Formatando a coluna datetime e definindo-a como √≠ndice para coluna para o conjunto de dados

```python
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
# Indexing Month
df = df.set_index('Month')
```

```python
df.head()
```

An√°lise Explorat√≥ria e Visualiza√ß√£o
Vamos come√ßar importando **matplotlib.pyplot e seaborn**

```python
import seaborn as sns
import matplotlib.pyplot
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Mapa de Calor

```python
#raw materiasl list
raw_data= ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price','Hard log Price', 'Hard sawnwood Price',
           'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
#Getting the correlation Matrix
Corrmat = df[raw_data].corr()
#Setting the size of plot
fig = plt.figure(figsize = (12, 9))
#masking the upper traingle part since matrix is symmetric (repetitive)
mask = np.triu (np.ones_like(Corrmat, dtype=bool))
sns.heatmap(Corrmat, vmax = .8, mask=mask, square = True, annot = True)
plt.show()
```

O mapa de calor mostra a correla√ß√£o entre as mat√©rias-primas, maior o valor correlacionado maior chance de serem duas mat√©rias-primas relacionadas, mas n√£o necessariamente


```python
plt.figure(figsize=(30,15))
changelist=['Copra price % Change', 'Softlog price % Change', 'Rubber price % Change', 'Cotton price % Change', 'Coarse wool price % Change',
'Fine wool price % Change', 'Hard log price % Change', 'Hard sawnwood price % Change']
#Generate a correlation Matrix for the whole dataset
corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

A partir desse gr√°fico podemos dizer que quase n√£o h√° rela√ß√£o entre a varia√ß√£o percentual dos pre√ßos das mat√©rias-primas:


*   O valor negativo implica que duas variaveis est√£o negativamente correlacionadas(um aumento outro decr√©scimo);
*   Zero implica nenhuma rela√ß√£o;
*   Caso contr√°rio, maior o valor maior a chance de rela√ß√£o pre√ßo e seus gr√°ficos de % de mudan√ßa.
